```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as sts
```

```python
# Leemos los datos normales de entrada
df = pd.read_csv("estaturas - normales.csv")
df
```

```python
# Aplicamos feature engineering para sustituir los valores NaN por el valor promedio
df = df.fillna(df.mean())
df
```

```python
sts.describe(df)
```

```python
# Calculamos la media del dataframe de entrada
u = df.mean()
u
```

```python
# Calculamos la varianza del dataframe
var = df.var()
var
```

```python
norm = sts.norm(u, var)
norm
```

```python
# Leemos los datos con anomalias
df_anomalias = pd.read_csv("estaturas - anomalias.csv")
df_anomalias
```

```python
# Definimos epsilon
e =  0.1

# Calculamos la probabilidad para los valores en las anomalias
for index, row in df.iterrows():
    p = norm.pdf(row)
    print(p.shape)
    print('La fila: ')
    print(row)
    print('Tiene una probabilidad de: ', p)
    if (p < e).any():
        # Es una anomalia
        print('Se trata de una ANOMALIA')
    else:
        print('Se trata de un valor NORMAL')
        
    print('\n')
```
