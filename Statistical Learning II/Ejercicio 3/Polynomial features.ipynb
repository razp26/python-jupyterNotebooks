# Conclusión

En las pruebas anteriormente realizadas, es posible ver la existencia de la llamada "maldición de la dimensionalidad", que a grandes rasgos, indica que a mayor número de variables o mayor complejidad en el modelo, habrá mayor dispersión entre los datos. Esto se traduce en que las distancias entre los datos dejarán de ser significativas. Esto implica que, para que las distancias entre los datos sean significativas, se requerirá de una gran cantidad de datos, lo que muchas veces no es posible conseguir.

Adicionalmente, si se contara con dicha cantidad de datos, el costo de procesamiento sería altísimo, lo cual haría que el tiempo de procesamiento fuera muy elevado.

Adicionalmente, cuando se tiene un número elevado de features, por lo general, la correlación existente es muy baja para la mayoría de ellas, a tal punto, que muchas veces puede decirse que no existe causalidad entre la variable a predecir y la mayoría de las features de entrada.

Lo recomendado es aplicar reducción de dimensionalidad para reducir la complejidad lo mayor posible y con ello evitar los tres problemas mencionados anteriormente.
